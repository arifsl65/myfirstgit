```python
import pandas as pd
```

```python
url= 'https://docs.google.com/spreadsheets/d/e/2PACX-1vSf2lI1ySZVYqgfY4hR56gx6_EomF9T6uS4Ji6RhkG53vmmwpqoHJ4Fmz-zDqgcoz_zgZu9G9p_LOtZ/pub?output=csv'
```

```python
df=pd.read_csv(url)
```

```python
type(df)
```

```python
df
```

```python
df.rename(columns={'Company Name': 'CompanyName'}, inplace=True)
```

```python
df.rename(columns={'Market Cap': 'MarketCap'}, inplace=True)
```

```python
df['Symbol'] = df['Symbol'].astype(str)
```

```python
df['CompanyName'] = df['CompanyName'].astype(str)
df['Industry'] = df['Industry'].astype(str)
```

```python
df['Industry'] = df['Industry'].astype(str)
```

```python
df['MarketCap'] = df['MarketCap'].astype(int)
```

```python
print(df)
```

```python
def add_single_quotes(value):
    return f"'{value}'"

# Apply the function to each element in the DataFrame
df_with_quotes = df.applymap(add_single_quotes)
```

```python

df = pd.DataFrame(df_with_quotes)

# Format the DataFrame with parentheses
formatted_df = df.apply(lambda col: f"('{col['Symbol']}', '{col['CompanyName']}', '{col['Industry']}', '{col['MarketCap']}')", axis=1)

# Print the formatted DataFrame
print(formatted_df)
```

```python
def format_row(row):
    return f"({row['Symbol']} , {row['CompanyName']} , {row['Industry']} , {row['MarketCap']})"

# Apply the function to each row using the apply method
df['FormattedData'] = df.apply(format_row, axis=1)

# Print the resulting DataFrame
print(df[['FormattedData']])
```

```python
df[['FormattedData']].to_csv('formatted_data.csv', index=False)
```
